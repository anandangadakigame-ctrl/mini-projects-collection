```python
def remove_outlier(col):
  sorted(col)
  Q1,Q3=col.quantile([0.25,0.75])
  IQR=Q3-Q1
  lower_range=Q1-(1.5*IQR)
  upper_range=Q3+(1.5*IQR)
  return lower_range, upper_range
```

```python
median1=df["rating"].median()
df["rating"].fillna(median1,inplace=True)
median2=df["user_age"].median()
df["user_age"].fillna(median2,inplace=True)
```

```python
mode1=df['app_category'].mode()[0]
df['app_category'].fillna(mode1,inplace=True)
mode2=df['app_name'].mode()[0]
df['app_name'].fillna(mode2,inplace=True)
mode3=df['review_language'].mode()[0]
df['review_language'].fillna(mode3,inplace=True)
```

```python
sns.distplot(df['rating'])
```

```python
sns.countplot(df['rating'],hue=df['app_name'])
```

```python
# Group by app_name and rating and count the occurrences
rating_counts = df.groupby(['app_name', 'rating']).size().reset_index(name='count')

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=rating_counts, x='rating', y='count', hue='app_name')
plt.title('Count of Ratings by App Name')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='App Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Display the correlation matrix
display(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
